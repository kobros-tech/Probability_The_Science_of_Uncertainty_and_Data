```python
# Jupyter / Python code
# ----------------------------

# Link failure probabilities
p_fail = [0.1, 0.05, 0.2]

# Success probability for each link
p_success = [1 - pf for pf in p_fail]

# 1) Probability that message reaches computer (any path works)
# Using independence and "at least one succeeds":
P_reach = 1 - p_fail[0] * p_fail[1] * p_fail[2]
print(f"Probability message reaches computer: {P_reach:.4f}")

# 2) Conditional probability if link1 fails (Bayes application)
# Define event B: message reaches computer
# Event A: link1 fails
# P(B | A) = probability message reaches given link1 fails
P_B_given_A = 1 - (1 - p_success[1]) * (1 - p_success[2])
print(f"P(message reaches | link1 fails): {P_B_given_A:.4f}")

# Suppose we want P(link1 fails | message reaches)
# Using Bayes rule:
P_A = p_fail[0]  # prior
P_B = P_reach     # total probability
P_A_given_B = (P_A * P_B_given_A) / P_B
print(f"P(link1 fails | message reaches): {P_A_given_B:.4f}")
```

```python
import matplotlib.pyplot as plt

# Example: Two-stage probability tree
# Stage 1: Event A occurs with probability 0.6, A complement 0.4
# Stage 2: Event B occurs given A: P(B|A)=0.7, P(B|A')=0.2

# Probabilities
P_A = 0.6
P_Ac = 0.4
P_B_given_A = 0.7
P_B_given_Ac = 0.2

# Total probability of B
P_B = P_A * P_B_given_A + P_Ac * P_B_given_Ac
print(f"Total probability of B: {P_B:.2f}")

# Tree diagram positions
fig, ax = plt.subplots(figsize=(8,6))
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
ax.axis('off')

# Stage 1 nodes
ax.text(0.5, 3, 'Start', ha='center', va='center', bbox=dict(boxstyle='round', facecolor='lightblue'))
ax.text(1.5, 4, 'A\nP=0.6', ha='center', va='center', bbox=dict(boxstyle='round', facecolor='lightgreen'))
ax.text(1.5, 2, "A'\nP=0.4", ha='center', va='center', bbox=dict(boxstyle='round', facecolor='lightcoral'))

# Stage 2 nodes
ax.text(3, 4.5, 'B\nP(B|A)=0.7', ha='center', va='center', bbox=dict(boxstyle='round', facecolor='yellow'))
ax.text(3, 3.5, "B'\nP(B'|A)=0.3", ha='center', va='center', bbox=dict(boxstyle='round', facecolor='orange'))
ax.text(3, 2.5, 'B\nP(B|A\')=0.2', ha='center', va='center', bbox=dict(boxstyle='round', facecolor='yellow'))
ax.text(3, 1.5, "B'\nP(B'|A')=0.8", ha='center', va='center', bbox=dict(boxstyle='round', facecolor='orange'))

# Draw arrows
ax.annotate('', xy=(1.4, 4), xytext=(0.6, 3), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(1.4, 2), xytext=(0.6, 3), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(3, 4.5), xytext=(1.6, 4), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(3, 3.5), xytext=(1.6, 4), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(3, 2.5), xytext=(1.6, 2), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(3, 1.5), xytext=(1.6, 2), arrowprops=dict(arrowstyle='->'))

plt.show()
```
