```python
# Sample space for two children
from itertools import product

children = ['B', 'G']
sample_space = list(product(children, repeat=2))
print("All possible outcomes (2 children):")
print(sample_space)
```

```python
# Filter outcomes to only those with at least one boy
king_possible = [outcome for outcome in sample_space if 'B' in outcome]
print("Outcomes with at least one boy:")
print(king_possible)
```

```python
# Simulate "have children until first boy" strategy with exactly 2 children
# Outcome is only possible if first child is girl, second child is boy
possible_outcome = ('G', 'B')
print(f"Under 'first boy' strategy, sibling is a girl: {possible_outcome[0] == 'G'}")
```

```python
def prob_sibling(sample_space, filter_condition, sibling_gender='G'):
    filtered = [outcome for outcome in sample_space if filter_condition(outcome)]
    sibling_count = sum(1 for outcome in filtered if sibling_gender in outcome)    
    return sibling_count / len(filtered)

# Classic 2-child fixed scenario
classic_prob = prob_sibling(sample_space, lambda outcome: 'B' in outcome)
print(f"Classic 2-child scenario: {classic_prob:.2f}")

# "Stop at first boy" scenario
first_boy_prob = prob_sibling(sample_space, lambda outcome: outcome == ('G','B'))
print(f"'Stop at first boy' scenario: {first_boy_prob:.2f}")
```
