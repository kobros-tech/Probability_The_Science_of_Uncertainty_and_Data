```python
# Import necessary libraries
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Define nodes
nodes = [
    "Real World\n(Data Generation)",
    "Statistics\n(Model Building)",
    "Probability Theory\n(Analysis Tools)",
    "Predictions & Decisions\n(Apply to Real World)"
]

# Add nodes to the graph
for node in nodes:
    G.add_node(node)

# Define edges (workflow)
edges = [
    ("Real World\n(Data Generation)", "Statistics\n(Model Building)"),
    ("Statistics\n(Model Building)", "Probability Theory\n(Analysis Tools)"),
    ("Probability Theory\n(Analysis Tools)", "Predictions & Decisions\n(Apply to Real World)")
]

# Add edges to the graph
G.add_edges_from(edges)

# Set layout
pos = nx.spring_layout(G, seed=42)  # positions for all nodes

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=4000, node_color="#f9f871", edgecolors="black")

# Draw edges
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=20, edge_color="black", width=2)

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold")

# Display plot
plt.title("Workflow: Real World  Statistics  Probability  Decisions", fontsize=12)
plt.axis("off")
plt.show()
```
